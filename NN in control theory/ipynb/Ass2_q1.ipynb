```python
# data-set generation
import numpy as np
from matplotlib import pyplot as plt
sfreq=1000
tmin = 0
tmax = 2
t = np.linspace(tmin,tmax,sfreq)
u = 0.4*np.sin(3*np.pi*t)
y= np.zeros(len(t))
for k in range(len(t)-1):
    y[k+1] = np.power(u[k],3)+y[k]/(1+y[k]**2)

# Visualization
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.scatter(t,u,s=10)
ax1.scatter(t,y,s=10)
ax1.set_title('Input u and output y versus time')
ax1.set_xlabel('t in seconds')
ax1.set_ylabel('Iutput Variable u/ Output variable y')
ax1.legend(['u','y'])
plt.show()    
# NN
```

```python
import keras
model = keras.Sequential()
# initialize the weights using he's method
model.add(keras.layers.Dense(10, input_dim=1, activation='relu', kernel_initializer='he_uniform'))
model.add(keras.layers.Dense(5, activation='relu', kernel_initializer='he_uniform'))
model.add(keras.layers.Dense(1))
model.compile(loss='mse', optimizer='adam')
model.fit(u, y, epochs=1000, batch_size=100, verbose=0)
yhat = model.predict(u)
```

```python

fig = plt.figure()
ax1 = fig.add_subplot(111)
plt.scatter(t,yhat, label='Predicted output',s=8)
plt.scatter(t,y, label='System output',s=8)
plt.title('predicted y versus Otime)')
plt.xlabel('t')
plt.ylabel('y)')
plt.legend()
plt.show()
```
